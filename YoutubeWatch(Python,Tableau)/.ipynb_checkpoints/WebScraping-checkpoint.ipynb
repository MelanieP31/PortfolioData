```python
#Video Id : youtube's URL after "v="
# index 50 'cause API-yt autorized only 50requests at a time
def get_video_details(youtube, video_ids):
    all_videos_stats = []
    
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part = 'snippet, statistics',
            id=','.join(video_ids[i:i+50]))
        response = request.execute()

        for video in response['items']:
            video_stats = dict(Title = video['snippet']['title'],
                                Published_date = video['snippet']['publishedAt'],
                                Views = video['statistics']['viewCount'],
                                Like = video['statistics']['likeCount'],
                                Comments = video['statistics']['commentCount'],
                                Videocategory = video['snippet']['categoryId'],
                              ChannelId = video['snippet']['channelId'],
                              ChannelTitle = video['snippet']['channelTitle']
                              )
            all_videos_stats.append(video_stats)

        return all_videos_stats
```

```python
#channel_id = 'UCOK0fZAUx82plnLhTKZW6qg'

channel_ids = ['UC7cs8q-gJRlGwj4A8OmCmXg',
'UCS2e0hEJMhwd6bNscS60xTg',
'UCj_iGliGCkLcHSZ8eqVNPDQ',
'UCDybamfye5An6p-j1t2YMsg'
]
```

```python
def get_channel_stats(youtube, channel_ids):
    all_data = []    
    request = youtube.channels().list(
        part="snippet,contentDetails,statistics,id",
        id = ','.join(channel_ids))
    response = request.execute()

    for i in range(len(response['items'])):

        data =dict(Channel_name = response['items'][i]['snippet']['title'],
                   Channel_id = response['items'][i]['id'], 
                   Channel_date = response['items'][i]['snippet']['publishedAt'],
                   Subscribers = response['items'][i]['statistics']['subscriberCount'],
                   Views = response['items'][i]['statistics']['viewCount'],
                   Total_videos = response['items'][i]['statistics']['videoCount'])
        
        all_data.append(data)
    
    return all_data
    
```

```python
channel_statitics = get_channel_stats(youtube, channel_ids)
```

```python
channel_data = pd.DataFrame(channel_statitics)
```

```python
channel_data['Subscribers'] = pd.to_numeric(channel_data['Subscribers'])
channel_data['Views'] = pd.to_numeric(channel_data['Views'])
channel_data['Total_videos'] = pd.to_numeric(channel_data['Total_videos'])
channel_data['Channel_id'] 
channel_data
```

# Stats descriptive

```python
#mean
mean_subscribers = numpy.mean(channel_data['Subscribers'])
mean_view = numpy.mean(channel_data['Views'])
mean_videos = numpy.mean(channel_data['Total_videos'])
print(f'mean-sub = {mean_subscribers}, mean_view = {mean_view}, mean_video = {mean_videos}')

#std
std_subscribers = numpy.std(channel_data['Subscribers'])
std_view = numpy.std(channel_data['Views'])
std_videos = numpy.std(channel_data['Total_videos'])
print(f'std-sub = {std_subscribers}, std_view = {std_view}, std_video = {std_videos}')

#loi de distribution
test_shapiro = scipy.stats.shapiro(channel_data['Total_videos'])
if test_shapiro.pvalue > 0.05:
    print(f'{test_shapiro} Loi normal')
else : 
    print(f'{test_shapiro} Pas loi normal')
```

# corrélation

```python
#corrélation Subscribers/views
def get_correlation(x,y) :
    pearson_r, pearson_p = scipy.stats.pearsonr(x,y)
    print(f'pearson_coeff_R = {pearson_r}, pearson_p-value = {pearson_p}')

    spearman_r = scipy.stats.spearmanr(x, y)
    print(f'spearman_coeff = {spearman_r}')

    slope, intercept, r, p, stderr = scipy.stats.linregress(x, y)
    print(f'linear_regression : slope:{slope}, intercept:{intercept}, coeffR:{r}, pvalue:{p}, stderr:{stderr}')
    line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'

    fig, ax = plt.subplots()
    ax.plot(x, y, linewidth=0, marker='s', label='Data points')
    ax.plot(x, intercept + slope * x, label=line)
    ax.set_xlabel(x.name)
    ax.set_ylabel(y.name)
    ax.legend(facecolor='white')
    
    plt.show()

get_correlation(channel_data['Subscribers'], y = channel_data['Views'])
get_correlation(channel_data['Subscribers'], y = channel_data['Total_videos'])
get_correlation(channel_data['Total_videos'], y = channel_data['Views'])
```
